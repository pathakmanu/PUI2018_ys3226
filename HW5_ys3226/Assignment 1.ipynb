# use groupby to count the building per LandUse code

```python
# group MN by land use
MN_grouped = MN.groupby(['LandUse'])['ZipCode'].count().reset_index()
MN_grouped.head()
```

```python
# group BK by land use
BK_grouped = BK.groupby(['LandUse'])['ZipCode'].count().reset_index()
BK_grouped.head()
```

# convert the landuse to integer

```python
MN_grouped['LandUse'] = MN_grouped['LandUse'].astype(int)
BK_grouped['LandUse'] = BK_grouped['LandUse'].astype(int)
MN_grouped.head() 
```

# merge the dataframes on land use

```python
both = pd.merge(BK_grouped, MN_grouped, on='LandUse')
both.columns
```

# rename things so that you have descriptive names

## rename the columns with descriptive names.

```python
both = both.rename(index=str, columns={"ZipCode_x": "bk", "ZipCode_y": "mn"})
both
```

## use the LandUse data dictionary and replace code w the verbose description of it

```python
# set up Land Use values to descriptive tag
data_dict = {1: "One & Two Family Buildings", 
             2: "Multi-Family Walk-Up Buildings",
             3: "Multi-Family Elevator Buildings",
             4: "Mixed Residential & Commercial Buildings",
             5: "Commercial & Office Buildings",
             6: "Industrial & Manufacturing",
             7: "Transportation & Utility",
             8: "Public Facilities & Institutions",
             9: "Open Space & Outdoor Recreation",
             10: "Parking Facilities",
             11: "Vacant Land"}
both = both.replace({"LandUse": data_dict})
both
```

# create a feature (a column, a series) with the ratio of number of buildings in BK vs MN and find out in which LandUse class Manhattan has more buildings than BK

```python
# calculate the ratio of BK to Man buildings per LandUse category
both['luratio'] = both['mn'] / both['bk']
both
```

```python
both['Manhattan wins'] = both['luratio'] > 1
both
```

## Comments

The table basically says two things: 

1- Brooklyn has much more land area than Manhattan does. That's why, the number of all kind of buildings except Multi-Family Elevator Buildings and other type of landuses are higher in Brooklyn.
