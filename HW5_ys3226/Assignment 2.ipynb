```python
SQL_SOURCE = 'https://fb55.carto.com/api/v2/sql?q=SELECT'

import urllib2
import urllib
import StringIO
import ast
import pandas as pd

def queryCartoDB(query, format='CSV', source=SQL_SOURCE):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    data = urllib.urlencode({'format': format, 'q': query})
    try:
        response = urllib2.urlopen(source, data)
    except urllib2.HTTPError, e:
        raise ValueError('\n'.join(ast.literal_eval(e.readline())['error']))
    except Exception:
        raise
    return response.read()
```

```python
task3_1 = '''
SELECT start_station_name, start_station_longitude, start_station_latitude
FROM fb55.citibike
GROUP BY start_station_name, start_station_longitude, start_station_latitude
'''
pd.read_csv(StringIO.StringIO(queryCartoDB(task3_1)), sep=',')
```

```python
#because it's a small urban area comparing the whole world, we can simplify the distance formula 
#and avoid sophisticated trigonometrics for this specific case
#distance formula in km: ((long1-long2)^2 + (lat1-lat2)^2)^0.5 * 6373 * 3.141592/180 
task3_2 = '''
SELECT DISTINCT start_station_name, start_station_longitude, start_station_latitude,
((start_station_longitude+73.9857)^2 + (start_station_latitude-40.7577)^2)^0.5 * 6373 * 3.141592/180 AS distance_km,
COUNT(cartodb_id) AS no_of_trips
FROM fb55.citibike
WHERE ((start_station_longitude+73.9857)^2 + (start_station_latitude-40.7577)^2)^0.5 * 6373 * 3.141592/180 < 0.5
GROUP BY start_station_name, start_station_longitude, start_station_latitude,
((start_station_longitude+73.9857)^2 + (start_station_latitude-40.7577)^2)^0.5 * 6373 * 3.141592/180
ORDER BY no_of_trips DESC
'''
pd.read_csv(StringIO.StringIO(queryCartoDB(task3_2)), sep=',')
```

```python
task4_1 = '''
SELECT start_station_name, start_station_longitude, start_station_latitude,
((start_station_longitude+73.9857)^2 + (start_station_latitude-40.7577)^2)^0.5 * 6373 * 3.141592/180 AS distance_km,
AVG(tripduration) AS avg_duration
FROM fb55.citibike
WHERE ((start_station_longitude+73.9857)^2 + (start_station_latitude-40.7577)^2)^0.5 * 6373 * 3.141592/180 < 0.5
AND (LEFT(TO_CHAR(starttime, 'DAY'), 5) = 'SATUR' OR LEFT(TO_CHAR(starttime, 'DAY'), 4) = 'SUND')
GROUP BY start_station_name, start_station_longitude, start_station_latitude, distance_km
ORDER BY avg_duration DESC
LIMIT 1
'''
pd.read_csv(StringIO.StringIO(queryCartoDB(task4_1)), sep=',')
```

```python
task4_Extra = '''
SELECT start_station_name, end_station_name, COUNT(cartodb_id) AS no_of_trips
FROM fb55.citibike
WHERE ((start_station_longitude+73.9857)^2 + (start_station_latitude-40.7577)^2)^0.5 * 6373 * 3.141592/180 < 0.5
AND tripduration < 60*60*2
GROUP BY start_station_name, end_station_name
ORDER BY no_of_trips DESC
'''
pd.read_csv(StringIO.StringIO(queryCartoDB(task4_Extra)), sep=',')
```
